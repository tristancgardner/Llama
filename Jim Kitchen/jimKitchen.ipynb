````python
def summarize_video_file(text):
    system_prompt = "You are a helpful assistant, an expert at summarizing text down to its most important and relevant information. You excel at undestanding an interviewee's main points as prompted by an interviewer's questions."

    user_prompt = f"""
    Summarize the {text} into it's most interesting, unique and/or important points. Each text is a transcription of a video file on a certain film day of an overall video production. 
    
    Use the following only as context: the speaker is Jim Kitchen - an entrepreneur and explorer who has gone to space as a civilian. 
    
    The summary should be no more than 6 sentences. Do not include descriptions of who Jim is as I've included in this prompt. Do not preceed the summary with any text like "Summary:" or "Here is the summary:" 
    """
    
    response = prompt_llama(model, tokenizer, system_prompt, user_prompt)       
    return response

data_new = data.copy()

for day_name, day_details in data_new.items():
    for file in day_details.get('Files', []):
        if file['File Summary'] == "" and file['Transcription'] != "":
            print(f'Llama summarizing file: {file["File"]} on day: {day_name}')
            response = summarize_video_file(file['Transcription'])
            file['File Summary'] = response
            print(f"File Summary: {response}\n")

display_dict_as_json(data_new)
````

````python
# now loop through the days and and summarize all the file summaries into a "Day Summary" for each day 
def get_day_summaries(data):
    all_day_summaries = ""
    for day_name, day_details in data.items():
        if day_details['Day Summary'] == "":
            day_summaries = "Film Day: " + day_name + "\n"
            for file in day_details.get('Files', []):
                day_summaries += "Interview file: " + file.get('File', '') + "\n"
                day_summaries += file.get('File Summary', '') + "\n"
            all_day_summaries += day_summaries + "\n"
    return all_day_summaries

all_day_summaries = get_day_summaries(data)
Markdown(all_day_summaries)
````

````python
def summarize_day_summaries(text):
    system_prompt = "You are a helpful assistant, an expert at summarizing text down to its most important and relevant information. You excel at undestanding the main points of speakers in interviews over one day or across multiple days of filming."

    user_prompt = f"""
    Summarize the {text} into it's most interesting, unique and/or important points and describe if there are any important relationships between files on the same day. 
    
    
    Give one "Day Summary" for each set of file/interview summaries separated by the title of the film day in the text. Try to capture all unique points of each file summary within the resulting day summary.
    If there are multiple days of file summaries in the text, also provide one "Film Shoot Summary" that describes the main points captured across the film days.
    """
    response = prompt_llama(model, tokenizer, system_prompt, user_prompt)
    return response

response = summarize_day_summaries(all_day_summaries)

Markdown(response)

# By new line characters, the text is formatted with the title of the film day, a file or 'interview' filmed on that day, and the resulting summary of what was said in the interview.
    
````
